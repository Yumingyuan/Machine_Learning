# 梯度下降法的向量化

```python
import numpy as np
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X=boston.data
y=boston.target
X=X[y<50.0]
y=y[y<50.0]
```

```python
from KNN.model_selection import train_test_split
X_train,y_train,X_test,y_test=train_test_split(X,y,random_seed=666)
```

```python
from PlayML.LinearRegression import LinearRegression
lin_reg1=LinearRegression()
%time lin_reg1.fit_normal(X_train,y_train)#正规方程方法
lin_reg1.score(X_test,y_test)
```

## 直接使用梯度下降法，有一些问题！！！

```python
lin_reg2=LinearRegression()
lin_reg2.fit_gd(X_train,y_train,eta=0.000001)
```

```python
lin_reg2.coef_
```

```python
lin_reg2.interception_
```

```python
lin_reg2.score(X_test,y_test)#分数比较低，说明默认的循环次数太少，没有找到合适的结果
```

```python
%time lin_reg2.fit_gd(X_train,y_train,eta=0.000001,n_iters=1e6)
```

```python
lin_reg2.score(X_test,y_test)#发现搜索过程太慢，需要进行归一化处理
```

## 使用梯度下降法前进行归一化，效果良好！

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardScaler=StandardScaler()
standardScaler.fit(X_train)
```

```python
X_train_standard=standardScaler.transform(X_train)
```

```python
lin_reg3 = LinearRegression()
%time lin_reg3.fit_gd(X_train_standard,y_train)
#归一化处理后，运行时间明显减少
```

```python
X_test_standard=standardScaler.transform(X_test)
```

```python
lin_reg3.score(X_test_standard,y_test)
```

## 梯度下降法的优势：特征个数非常大时比正规方程快

```python
m=1000#样本数
n=5000#特征数
big_X = np.random.normal(size=(m,n))
true_theta = np.random.uniform(0.0,100.0,size=n+1)#5001个theta值
true_theta.shape
big_y=big_X.dot(true_theta[1:])+true_theta[0]+np.random.normal(0.,10.)#噪声向量有m个元素，也就是给每一个样本的值都加上一个新生成的噪声
```

```python
big_y
```

```python
big_reg1=LinearRegression()
%time big_reg1.fit_normal(big_X,big_y)
```

```python
big_reg2=LinearRegression()
%time big_reg2.fit_gd(big_X,big_y)
```
